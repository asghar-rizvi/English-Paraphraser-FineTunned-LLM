# Importing Libraries

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM,DataCollatorForSeq2Seq
from datasets import load_dataset
from peft import LoraConfig, get_peft_model, TaskType
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
import random
import numpy
```

# Importing t5 small

```python
base_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

# Preprocess Data

```python
def preprocess_data(example):
    # Add diversity by random prefix
    prefixes = ["paraphrase: ", "rewrite: ", "rephrase: "]
    prefix = random.choice(prefixes)

    inputs = prefix + example["sentence1"]
    targets = example["sentence2"]

    model_inputs = tokenizer(
        inputs,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    labels = tokenizer(
        targets,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    return {
        "input_ids": model_inputs["input_ids"],
        "attention_mask": model_inputs["attention_mask"],
        "labels": labels["input_ids"]
    }
```

```python
dataset = dataset.map(preprocess_data)
```

```python
dataset
```

```python
train = dataset["train"].remove_columns(["id", "sentence1", "sentence2", "label"])
validation = dataset["validation"].remove_columns(["id", "sentence1", "sentence2", "label"])
test = dataset["test"].remove_columns(["id", "sentence1", "sentence2", "label"])
```

# Lora Configuration

```python
lora_config = LoraConfig(
    r = 16,
    lora_alpha = 32,
    lora_dropout = 0.1,
    task_type = TaskType.SEQ_2_SEQ_LM,
    target_modules=["q", "v", "k", "o", "wi", "wo"],
    bias = 'none'
)
```

```python
model = get_peft_model(base_model, lora_config)
```

## Difference Between Normal Training And Lora

```python
model.print_trainable_parameters()
```

 # Building Trainer

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./paraphrase_model",
    eval_steps=500,
    save_steps=500,
    learning_rate=5e-4,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    predict_with_generate=True,
    generation_max_length=128,
    warmup_steps=500,
    fp16=True,
    metric_for_best_model="rougeL",
    greater_is_better=True,
    label_names = ['labels']
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train,
    eval_dataset=validation,
    tokenizer=tokenizer,
    data_collator=DataCollatorForSeq2Seq(tokenizer, model=model)
)
```

## Fine Tunning LLM

```python
trainer.train()
```

# Results Of Fine Tunning

```python
results = trainer.evaluate()
print("Evaluation Results:", results)
```

# Testing Fine Tunned LLM On Unseen Data

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from peft import PeftModel, PeftConfig

peft_model_id = "./final_paraphraser"
config = PeftConfig.from_pretrained(peft_model_id)
base_model = AutoModelForSeq2SeqLM.from_pretrained(config.base_model_name_or_path)
model = PeftModel.from_pretrained(base_model, peft_model_id)
tokenizer = AutoTokenizer.from_pretrained(peft_model_id)

model = model.merge_and_unload()
```

```python
def paraphrase(
    text: str,
    num_return_sequences: int = 3,
    temperature: float = 0.9,
    max_length: int = 128
) -> list:
    prompts = [
        "Paraphrase this technically: ",
        "Rewrite this formally: ",
        "Rephrase for clarity: ",
        "Express this differently: "
    ]
    prompt = random.choice(prompts)

    inputs = tokenizer(
        prompt + text,
        return_tensors="pt",
        truncation=True,
        max_length=max_length,
        padding="max_length"
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_length=max_length,
        do_sample=True,
        temperature=temperature,
        top_k=50,
        top_p=0.95,
        num_return_sequences=num_return_sequences,
        repetition_penalty=2.5,
        no_repeat_ngram_size=3
    )

    paraphrases = []
    for output in outputs:
        decoded = tokenizer.decode(output, skip_special_tokens=True)

        # Remove any leftover prompt
        for p in prompts:
            decoded = decoded.replace(p, "")
        paraphrases.append(decoded.strip())

    return paraphrases
```

## Example 1

```python
text = """The rapid development of artificial intelligence has raised significant ethical concerns,
particularly regarding data privacy and algorithmic bias. Many experts argue that comprehensive
regulations need to be established before these technologies become more pervasive in society."""

paraphrases = paraphrase(text)

print("Original:\n", text)
print("\nParaphrases:")
for i, para in enumerate(paraphrases, 1):
    print(f"\n{i}. {para}")
```
